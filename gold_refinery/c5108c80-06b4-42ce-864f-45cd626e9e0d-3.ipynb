### О проекте

***Описание данных***

**Технологический процесс**:
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant 
- Xanthate — ксантогенат (промотер, или активатор флотации);
- Sulphate — сульфат (на данном производстве сульфид натрия);
- Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

**Параметры этапов**:
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

**Наименование признаков**

Наименование признаков должно быть такое: `[этап].[тип_параметра].[название_параметра]` *(Пример: rougher.input.feed_ag)*

**Возможные значения для блока [этап]**:
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики

**Возможные значения для блока [тип_параметра]**:
- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

### Проверка вычисления эффективности обогащения 

Эффективность обогащения вычисляется по формуле:

$\ Recovery = \frac{C \times (F-T)}{F \times (C-T)} \times 100%$, где:
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.


Проверим правильность рассчета эффективности отчистки на этапе флотации. Рассчитаем эффективность по формуле выше и сравним с имеющимися значениями, определив среднюю абсолютную ошибку для получившихся и данных значений.

```python
def recovery_rate(c, f, t):
    recovery = (c * (f - t)) / (f * (c - t)) * 100
    return recovery
```

```python
calculated_recovery = recovery_rate(
    data_train['rougher.output.concentrate_au'], 
    data_train['rougher.input.feed_au'], 
    data_train['rougher.output.tail_au']
)

mean_absolute_error(
    calculated_recovery,
    data_train['rougher.output.recovery']
)
```

Средняя абсолютная ошибка стремится к 0, в связи с чем можно сделать вывод о правильности данных в тренировочной выборке.

### Анализ изменения концентрации элементов на различных этапах

Рассмотрим изменение концентрации элементов Au, Ag, Pb на каждом этапе.

```python
au_concentrate = { 
    'rougher.input.feed_au': 'Сырье', 
    'rougher.output.concentrate_au' : 'Флотация', 
    'primary_cleaner.output.concentrate_au' : 'Первичная очистка',
    'final.output.concentrate_au' : 'Финальный концентрат'
}

ag_concentrate = {
    'rougher.input.feed_ag': 'Сырье', 
    'rougher.output.concentrate_ag' : 'Флотация', 
    'primary_cleaner.output.concentrate_ag' : 'Первичная очистка',
    'final.output.concentrate_ag' : 'Финальный концентрат'
}

pb_concentrate = {
    'rougher.input.feed_pb': 'Сырье', 
    'rougher.output.concentrate_pb' : 'Флотация', 
    'primary_cleaner.output.concentrate_pb' : 'Первичная очистка',
    'final.output.concentrate_pb' : 'Финальный концентрат'
}

def stages_concentration(element, stages):
    sns.set(style="darkgrid")
    for parameter, stage in stages.items():
        sns.histplot(
            data=data_full, 
            x=parameter,
            label=stage
        )
    plt.title(element)
    plt.xlabel('Концентрация')
    plt.legend() 
    plt.show()
```

```python
stages_concentration('Золото', au_concentrate)
```

Как мы видим, на всех стадиях очистки, концентрация золота равномерно растет с каждой стадией, что логично, с учетом того, что это анализ данных золотодобывающей компании, и основным добывающимся полезным ископаемым является золото.

```python
stages_concentration('Серебро', ag_concentrate)
```

Анализ распределения значений серебра показывает, что количество серебра уменьшается с каждой стадии очистки, при этом содержание серебра в исходном сырье не превышает значений на какой-либо из стадии, кроме финального концентрата. Это можно объяснить тем, что во время флотации в первую очередь порода очищается от пустой породы, поэтому после флотации концентрация серебра растет, и снижается лишь после первичной и вторичной очистки.

```python
stages_concentration('Свинец', pb_concentrate)
```

Можно отметить, что распределение значений концентрации свинца сильно отличается от серебра и ближе к изменению распределений концентрации золота.
В изначальном сырье концентрация свинца гораздо ниже, чем на остальных стадиях очистки, после флотации заметно повышается, а после первичной и вторичной очистки, наиболее частые значения концентраций схожи, меняются лишь края распределений. Это может быть связано с тем, что свинец при высоких температурах поглощает примеси, которые имеются у других металлов, из-за чего его концентрация растет. Другая возможная причина - сохранение свинца для последующей переработки с целью добычи дополнительного золота (в свинце может содержаться до 40% золота).

### Суммарные концентрации металлов

### Создание оценочной метрики 

Напишем функцию для определения оценочной метрики sMAPE

```python
def calculate_smape(target, predictions):
    target_values = target.values if isinstance(target, pd.DataFrame) else target
    predictions_values = predictions.values if isinstance(predictions, pd.DataFrame) else predictions
    smape_rougher = (2 * np.abs(predictions[:, 0] - target_values[:, 0]) / (np.abs(predictions[:, 0]) + np.abs(target_values[:, 0]))) * 100
    smape_final = (2 * np.abs(predictions[:, 1] - target_values[:, 1]) / (np.abs(predictions[:, 1]) + np.abs(target_values[:, 1]))) * 100
    
    smape = 0.25 * np.mean(smape_rougher) + 0.75 * np.mean(smape_final)
    
    return smape

scorer = make_scorer(calculate_smape, greater_is_better=False)
```

- смасштабируем признаки
